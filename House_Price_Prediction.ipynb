```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
#Feeding the data set
data=pd.read_csv("homeprices.csv")
print(data)
```

```python
#Creating the plots to check if the dataset suits a regression model
plt.scatter(data.area,data.price, color="orange",marker="+")
plt.xlabel("Area - Sq.ft")
plt.ylabel("Price in INR")
plt.title("HOUSE PRICE PREDICTION")

plt.figure()
plt.plot(data.area,data.price)
plt.xlabel("Area - Sq.ft")
plt.ylabel("Price in INR")
plt.title("HOUSE PRICE PREDICTION")

plt.figure()
plt.scatter(data.area,data.price, color="orange",marker="+")
plt.plot(data.area,data.price)
plt.xlabel("Area - Sq.ft")
plt.ylabel("Price in INR")
plt.title("HOUSE PRICE PREDICTION")


```

```python
#Creation of an object
reg= linear_model.LinearRegression()
#Fitting the model
reg.fit(data[["area"]], data.price)
```

```python
#Making predictions
reg.predict([[3300]])
```

```python
#To find the intercept c in eqn. y=mx+c
reg.intercept_
#It is the beginning of the minimum price fetched from the model
```

```python
#To find the coefficient/slope m in eqn. y=mx+c
reg.coef_
```

```python
#Formula
#y=mx+c where y=price, x=area
price=135.78767123*3300+180616.43835616432
print(price)
```

```python
#Loading the 2nd dataset as a whole
df=pd.read_csv("areas.csv")
print(df)
```

```python
#Predict price for given dataset
prices=reg.predict(df)
```

```python
#Creating a column with the name 'prices'
df['prices']=prices
df
```

```python
#Exporting the latest csv file
df.to_csv("areas_vs_price.csv",index=False)
```

```python
#Creation of plots for the 2nd dataset
plt.scatter(df.area,df.prices)
plt.plot(df.area, reg.predict(df[['area']]), color='orange')
plt.title("Area v/s Price Prediction for Housing")
plt.xlabel("Area")
plt.ylabel("Price")
```
